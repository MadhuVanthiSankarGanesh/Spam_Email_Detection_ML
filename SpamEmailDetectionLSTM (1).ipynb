```python
def get_email_content(email_path):
    file = open(email_path,encoding='latin1')
    try:
        msg = email.message_from_file(file)
        for part in msg.walk():
            if part.get_content_type() == 'text/plain':
                return part.get_payload() # prints the raw text
    except Exception as e:
        print(e)


def get_email_content_bulk(email_paths):
    email_contents = [get_email_content(o) for o in email_paths]
    return email_contents
```

```python
import re
import string
from nltk.tokenize import word_tokenize
```

```python
def remove_hyperlink(word):
    return  re.sub(r"http\S+", "", word)
```

```python
def to_lower(word):
    result = word.lower()
    return result
```

```python
def remove_number(word):
    result = re.sub(r'\d+', '', word)
    return result
```

```python
def remove_punctuation(word):
    result = word.translate(str.maketrans(dict.fromkeys(string.punctuation)))
    return result
```

```python
def remove_whitespace(word):
    result = word.strip()
    return result
```

```python
def replace_newline(word):
    return word.replace('\n','')
```

```python
def clean_up_pipeline(sentence):
    cleaning_utils = [remove_hyperlink,
                      replace_newline,
                      to_lower,
                      remove_number,
                      remove_punctuation,remove_whitespace]
    for o in cleaning_utils:
        sentence = o(sentence)
    return sentence
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer


x_train_flat = [' '.join(x) for x in x_train]
x_test_flat = [' '.join(x) for x in x_test]

# Convert text data to numerical features using CountVectorizer
vectorizer = CountVectorizer(max_features=200)
x_train_features = vectorizer.fit_transform(x_train_flat)
x_test_features = vectorizer.transform(x_test_flat)

import tensorflow as tf
from tensorflow import keras
import re
import nltk
from nltk.corpus import stopwords
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Set the sentence length and vocabulary size
sentence_len = 200
vocab_size = 10000

# Tokenize text data using Keras Tokenizer
tokenizer = Tokenizer(num_words=vocab_size, oov_token="<OOV>")
tokenizer.fit_on_texts(x_train_flat)

# Convert text data to sequences
x_train_sequences = tokenizer.texts_to_sequences(x_train_flat)
x_test_sequences = tokenizer.texts_to_sequences(x_test_flat)

# Pad sequences to have a consistent length
x_train_padded = pad_sequences(x_train_sequences, maxlen=sentence_len, padding="post", truncating="post")
x_test_padded = pad_sequences(x_test_sequences, maxlen=sentence_len, padding="post", truncating="post")

# Define a neural network model with Bidirectional LSTM
model = keras.Sequential([
    keras.layers.Input(shape=(sentence_len,)),  # Input layer with shape matching sentence length
    keras.layers.Embedding(input_dim=vocab_size, output_dim=128),  # Embedding layer
    keras.layers.Bidirectional(keras.layers.LSTM(100, return_sequences=True)),  # Bidirectional LSTM layer
    keras.layers.GlobalMaxPooling1D(),  # Global max-pooling layer
    keras.layers.Dense(128, activation='relu'),  # Additional dense layer
    keras.layers.Dense(1, activation='sigmoid')  # Output layer (binary classification)
])

# Compile the model with a lower learning rate and early stopping
optimizer = keras.optimizers.Adam(learning_rate=0.0001)
model.compile(optimizer=optimizer, loss='binary_crossentropy', metrics=['accuracy'])

# Implement early stopping based on validation loss
early_stopping = keras.callbacks.EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model with the modified configurations
model.fit(x_train_padded, y_train, epochs=100, batch_size=64, validation_split=0.2, callbacks=[early_stopping])

# Evaluate the model
loss, accuracy = model.evaluate(x_test_padded, y_test)
print(f'Test accuracy: {accuracy}')
print(f'Test loss: {loss}')
```

```python
import nltk
import re
import numpy as np
from nltk.corpus import stopwords
from tensorflow.keras.preprocessing.text import one_hot
from tensorflow.keras.preprocessing.sequence import pad_sequences

# Define a function to classify messages from a text file
def classify_messages_from_file(model, file_path):
    # Read the text from the file
    with open(file_path, 'r', encoding='utf-8', errors='ignore') as file:
        text = file.read()

    # Tokenize the input text into individual sentences
    sentences = nltk.sent_tokenize(text)

    # Initialize variables to count the number of spam and non-spam sentences
    num_spam = 0
    num_non_spam = 0

    for sentence in sentences:
        # Clean up the sentence using various text preprocessing steps
        cleaned_sentence = clean_up_pipeline(sentence)

        # Perform one-hot encoding for the cleaned sentence
        one_hot_encoded = one_hot(cleaned_sentence, vocab_size)

        # Pad or truncate the sequence to have a consistent length (sentence_len)
        text = pad_sequences([one_hot_encoded], maxlen=sentence_len, padding="pre", truncating="pre")

        # Predict the text using the model
        predict = model.predict(text)

        # If the predicted value is greater than 0.5, it's a spam; otherwise, it's not a spam
        if predict > 0.5:
            num_spam += 1
        else:
            num_non_spam += 1

    # Determine the overall result based on the majority of sentences
    if num_spam > num_non_spam:
        return "Overall Result: It is a spam"
    else:
        return "Overall Result: It is not a spam"

# Example file path
file_path = "/content/data2/spam_data/spam/00001.7848dde101aa985090474a91ec93fcf0"  # Replace with the path to your text file

# Call the function to classify messages from the text file
result = classify_messages_from_file(model, file_path)

# Print the overall result
print(result)
```

```python
def plot_confusion_matrix(cm, classes,
                          normalize=False,
                          title='Confusion matrix',
                          cmap=plt.cm.Blues):
    """
    This function prints and plots the confusion matrix.
    Normalization can be applied by setting `normalize=True`.
    """
    if normalize:
        cm = cm.astype('float') / cm.sum(axis=1)[:, np.newaxis]
        print("Normalized confusion matrix")
    else:
        print('Confusion matrix, without normalization')

    print(cm)

    plt.imshow(cm, interpolation='nearest', cmap=cmap)
    plt.title(title)
    plt.colorbar()
    tick_marks = np.arange(len(classes))
    plt.xticks(tick_marks, classes, rotation=45)
    plt.yticks(tick_marks, classes)

    fmt = '.2f' if normalize else 'd'
    thresh = cm.max() / 2.
    for i, j in itertools.product(range(cm.shape[0]), range(cm.shape[1])):
        plt.text(j, i, format(cm[i, j], fmt),
                 horizontalalignment="center",
                 color="white" if cm[i, j] > thresh else "black")

    plt.ylabel('True label')
    plt.xlabel('Predicted label')
    plt.tight_layout()
```

```python
# Plot normalized confusion matrix
plt.figure()
plot_confusion_matrix(cnf_matrix, classes=['Non Spam','Spam'], normalize=False,
                      title='Confusion matrix')
```
